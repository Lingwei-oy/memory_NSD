```python
from cloudpathlib import S3Path, S3Client
from pathlib import Path
from utilities import ls, crawl
import nibabel as nib

# Set up our cache path:
cache_path = Path('/home/jovyan/cache')
if not cache_path.exists():
    cache_path.mkdir()

# Create the root S3Path for the NSD:
nsd_base_path = S3Path(
    's3://natural-scenes-dataset/',
    client=S3Client(
        no_sign_request=True,
        local_cache_dir=cache_path))
```

```python
import numpy as np
from nilearn.image import new_img_like

# Step 1: Set up the path to the MGZ file
aseg_path = nsd_base_path / 'nsddata' / 'freesurfer' / 'subj01' / 'mri' / 'aparc.a2009s+aseg.mgz'

# Step 2: Load the .mgz file directly using nibabel
aseg_img = nib.load(aseg_path.fspath)         # .fspath returns the locally cached version
aseg_data = aseg_img.get_fdata()

# Step 3: Define prefrontal label indices from Destrieux atlas
# ⚠️ These are placeholder values — you’ll want to update them based on the .ctab or FSColorLUT
pfc_label_indices = [
    1,   # G&S_frontomargin
    5,   # G&S_transv_frontopol
    13,  # G_front_inf-Opercular
    14,  # G_front_inf-Orbital
    15,  # G_front_inf-Triangul
    16,  # G_front_middle
    17,  # G_front_sup
    24,  # G_orbital
    31,  # G_rectus
]

# Step 4: Create a binary mask for the PFC regions
pfc_mask_data = np.isin(aseg_data, pfc_label_indices).astype(np.uint8)

# Step 5: Create a new NIfTI image in memory (no saving needed)
pfc_mask_img = new_img_like(aseg_img, pfc_mask_data)
```

```python
from nilearn import plotting

plotting.plot_roi(pfc_mask_img, bg_img=aseg_img, title="Prefrontal Cortex Mask")
```
